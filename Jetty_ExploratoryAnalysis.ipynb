```python
df_jetties_nulls = df_jetties.isnull().sum().sort_values() # check to see how many null objects
df_jetties_nulls = df_jetties_nulls.to_frame()
df_jetties_nulls.rename(columns = {0: 'Null_Count'}, inplace = True)
df_jetties_nulls_groups = df_jetties_nulls.groupby(['Null_Count']).size()
df_jetties_nulls_groups = df_jetties_nulls_groups.to_frame().reset_index()
df_jetties_nulls_groups.rename(columns = {0 : 'Missing_Values_Count'}, inplace = True)
# computing proportions of nulls (as percentages)
total_columns = df_jetties.shape[1];
df_jetties_nulls_groups['Missing_Values_Count']  = df_jetties_nulls_groups['Missing_Values_Count']  / total_columns * 100
df_jetties_nulls_groups_sorted = df_jetties_nulls_groups.sort_values(by='Missing_Values_Count', ascending = False)
df_jetties_nulls_groups_sorted = df_jetties_nulls_groups_sorted.set_index('Null_Count')
df_jetties_nulls_groups_sorted
```

```python
df_jetties_unique = df_jetties.nunique(axis=0).sort_values() # count unique values in each column
df_jetties_unique = df_jetties_unique.to_frame()
df_jetties_unique.rename(columns = {0: 'Unique_Count'}, inplace = True)
df_jetties_unique_group = df_jetties_unique.groupby(['Unique_Count'], as_index = False)
df_jetties_unique_group = df_jetties_unique['Unique_Count']

# Create df_jetties_ids dataframe specified by id locations
df_jetties_ids = df_jetties_unique.loc[['asset_id', 'photo_id_p', 'polygon_id','pathway_id', '_id']]
list_ids = []
for x in df_jetties_ids['Unique_Count'].to_list():
    if x == 0:
        list_ids.append('Empty')
    if x == 1:
        list_ids.append('Duplicates')
    if x == df_jetties.shape[0]:
        list_ids.append('Unique')

df_jetties_ids['Id_Status'] = list_ids
df_jetties_ids
```

## Data Visualization

```python
# Visualising proportions of material_1 attribute used across Jetty assets
df_material_1 = df_jetties.material_1.value_counts()
df_material_1 = df_material_1.to_frame()
df_material_1.info()
```

```python
colors_list = ['gold', 'yellowgreen', 'lightcoral', 'lightskyblue', 'lightgreen', 'pink', 'red']
explode_list = [0.1, 0, 0, 0, 0.1, 0.1, 0.1] # ratio for each continent with which to offset each wedge.

df_material_1['material_1'].plot(kind='pie',
                            figsize=(15, 6),
                            autopct='%1.1f%%', 
                            startangle=90,    
                            shadow=True,       
                            labels=None,         # turn off labels on pie chart
                            pctdistance=1.12,    # the ratio between the center of each pie slice and the start of the text generated by autopct 
                            colors=colors_list,  # add custom colors
                            explode=explode_list # 'explode' lowest 3 continents
                            )

# scale the title up by 12% to match pctdistance
plt.title('Material_1 Data Proportion across Jetty Assets', y=1.12) 

plt.axis('equal') 

# add legend
plt.legend(labels=df_material_1.index, loc='upper left') 
plt.savefig('material_1_chart.png', bbox_inches='tight')
plt.show()
```

```python
# Visualising Proportions of Missing Values
# plotting bar chart
ax=df_jetties_nulls_groups_sorted.plot(kind='bar', figsize=(20, 8), width=0.8, 
                                       color=('#5cb85c', '#5bc0de', '#d9534f', '#DFFF00', '#FFBF00', '#DE3163', '#CCCCFF'), 
                                       fontsize=14) 
ax.set_title("Percentage of Column Attributes per Missings Values Count", fontsize=16)
ax.legend(fontsize=14)
ax.spines['left'].set_visible(False) # Remove specified borders
ax.spines['top'].set_visible(False) # Remove specified borders
ax.spines['right'].set_visible(False) # Remove specified borders
ax.axes.get_yaxis().set_visible(False) # Turn off y-axis
# find the values and append to list
# get_x() and get_height() to obtain the coordinates for the percentage values.
y_data = []
x_data = []
x_offset=0.03
y_offset=0.5
for i in ax.patches:
    y_data.append(i.get_height())
    x_data.append(i.get_x())
    ax.annotate(str(i.get_height().round(1))+ '%',(i.get_x()+x_offset, i.get_height()+y_offset))
plt.savefig('missing_values_barchart.png', bbox_inches='tight')   
```

```python
# Asset Condition and Material_1 & Material_2
df_jetties_group = df_jetties[['material_1','material_2','condition']]

# Replace NaN in material_2 with same type as material_1
for k in range(df_jetties.shape[0]):
    if str(df_jetties['material_2'][k]) == 'nan':
        df_jetties['material_2'][k] = (df_jetties['material_1'][k])
    
grouped_jetties = df_jetties_group.groupby(['material_1','material_2'], as_index = False).mean()

# jetties_group pivot and fill missing values with 0
grouped_jetties_pivot = grouped_jetties.pivot(index = 'material_1', columns = 'material_2').fillna(0)
grouped_jetties_pivot
```

```python
#use the grouped results to generate a heatmap that plots the target variable (condition) proportional to colour with
# respect to the variables 'material_1' and 'material_2'
# heatmap with default labelling
plt.pcolor(grouped_jetties_pivot, cmap='RdYlBu_r')
plt.colorbar()
plt.xlabel('material_2')
plt.ylabel('material_1')
plt.savefig('condition_heatmap_default.png', bbox_inches='tight')
plt.show()
```

```python
# adding some custom labelling to convey useful information
fig, ax = plt.subplots()
im = ax.pcolor(grouped_jetties_pivot, cmap='RdYlBu_r')

#label names
row_labels = grouped_jetties_pivot.columns.levels[1]
col_labels = grouped_jetties_pivot.index

#move ticks and labels to the center
ax.set_xticks(np.arange(grouped_jetties_pivot.shape[1]) + 0.5, minor=False)
ax.set_yticks(np.arange(grouped_jetties_pivot.shape[0]) + 0.5, minor=False)

#insert labels
ax.set_xticklabels(row_labels, minor=False)
ax.set_yticklabels(col_labels, minor=False)

#rotate label if too long
plt.xticks(rotation=90)

fig.colorbar(im)
plt.savefig('condition_heatmap_detailed.png', bbox_inches='tight')
plt.show()
```

## Conclusion and Recommendations
The exploratory data analysis highlighted notable data quality issues (inconsistencies) and some initial insights gained from the provided data. The below table lists some of the major issues and the recommendations to mitigate such data quality issues.

| Data Quality Issues | Description (Comments)    | Recommendation (Mitigation) |
|---------------------|---------------------------|-----------------------------|
| Missing Values      | Various columns have empty values, zeros or dates in certain/entire records e.g., __asset_id__, __height, depth, width, length, area, _modified__ | Check and fill the records. Else filter out the record if not critical|
| Duplicates| Some columns have Id records assigned all to zeros e.g., __pathway_id__, __polygon_id__  | Check and fill Ids correctly and preserve uniqueness where required|
|Inconsistent or incomplete columns naming | For example, '__field_irri__' and __created_b__ |Apply clear and consistent columns naming convention |
| Column Attributes Description | Attributes description is missing | Provide a concise description of each attribute. This information is crucial when trying to analyse and make sense of the data|
| Geospatial Data | GPS coordinates for assets identified by location names e.g., __parkreserv__ | Provide GPS coordinates to overlay some of the features on the map|
